```python
file1=open('2348.txt','r')
```

```python
file_buffer=file1.read()
```

```python
file_buffer
```

```python
file_list=file_buffer.split()
```

```python
file_list
```

```python
import collections as cc
```

```python
file_collection=cc.Counter(file_list)
```

```python
file_collection
```

```python
file_dict=dict(file_collection)
```

```python
x,y,z=map(int, input().split())
```

```python
file_word_features=[(i,j) for i,j in file_dict.items() if(len(i)>x)and(j>y)and(j<z)]
```

```python
file_word_features
```

```python
file_feature_list=str(file_word_features)
```

```python
file2=open('feature_list.txt','w')
```

```python
file2.write(file_feature_list)
```

```python
file1.close()
```

```python
file2.close()
```
